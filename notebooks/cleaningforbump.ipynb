```python
import pandas as pd
```

```python
df = pd.read_csv("../data/Unicorn_Companies_clean.csv")
```

```python
df
```

```python
df = df.loc[df["Select Inverstors"] != 'None']
```

```python
#remove dollar sign
# data.loc[0:2,['Num','NAME']] = [100,'Python']
df["Valuation ($B)"] = df["Valuation ($B)"].str.strip("$")
df["Valuation ($B)"].astype(float)
```

```python
df.loc[df["Industry"] == "Finttech"]["Industry"] = "Fintech"
```

```python
df['Industry'] = df['Industry'].replace(['Finttech'],'Fintech')
df["Industry"] = df["Industry"].replace(['Artificial intelligence'], 'Artificial Intelligence')
```

```python
df["Date Joined"] = pd.to_datetime(df["Date Joined"])
```

```python
df['YearJoined'] = pd.DatetimeIndex(df["Date Joined"]).year
```

```python
df
```

```python
df_count = df.groupby(["Industry", "YearJoined"]).count()
```

```python
df_count["NumberOfUnicorns"] = df_count["Unnamed: 0"] 
```

```python
df_count = df_count[['NumberOfUnicorns']]
```

```python
df_count['index1'] = df_count.index
```
